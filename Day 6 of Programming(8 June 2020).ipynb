```python
#importing part
#Matplotlib today

import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sb
```

```python
df = pd.read_csv('E:\Kaiwalya\Courses\CodeQuest\Machine Learning using python material\Data sets\pokemon.csv', index_col=0)#identify column 0 as id
```

```python
df.head()
```

```python
#generating bar chart with generation id
sb.countplot(data = df, x = 'generation_id')
```

```python
#genrate bar chart with generation_id with specific colour

base_color = sb.color_palette()[4]
sb.countplot(data = df, x = 'generation_id',color = base_color)
```

```python
#generate Bar chart from highest to lower value
cat_order = df['generation_id'].value_counts().index
sb.countplot(data = df, x ='generation_id', color = base_color, order = cat_order)
```

```python
#generate horizontal bar chart
base_color = sb.color_palette()[0]
sb.countplot(data = df, y = 'generation_id', color = base_color)
```

```python
#generate pie chart for generation id

sorted_counts = df['generation_id'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False)
```

```python
#generate doughnut chart for generation id

sorted_counts = df['generation_id'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, wedgeprops = {'width':0.7});
plt.axis('square')
```

```python
plt.hist(data = df, x = 'attack')
```

```python
bin_edges = np.arange(0, df['attack'].max()+1, 1)
plt.hist(data = df, x = 'attack', bins = bin_edges)
plt.show()
```

```python
plt.figure(figsize = [10, 5])#Larger figure size for subplots

#histogram on left, example of too-large bin size
plt.subplot(1,2,1)#i row, 2 cols, subplot 1

bin_edges = np.arange(0, df['attack'].max()+4, 4)
plt.hist(data = df, x = 'attack', bins = bin_edges)

#histogram on righ, example of too-small bin size
plt.subplot(1,2,2) #1 row , 2cols, suplot 2

bin_edges2 = np.arange(0, df['attack'].max()+0.5, 0.5)
plt.hist(data = df, x = 'attack', bins = bin_edges2)
plt.show()
```

```python
plt.scatter(data = df, x = 'attack', y = 'defence')
```

```python
plt.figure(figsize = [12, 5])
#left plot: scatterplot of discrete data with jitter and transparecy

plt.subplot(1, 2, 1)
sb.regplot(data = df, x = 'height', y = 'weight', fit_reg = False, x_jitter = 0.2, y_jitter = 0.2, scatter_kws = {'alpha':1/3})

#right plot : heat map with bin edges between values

plt.subplot(1, 2, 2)
bins_x = np.arange(0.5, 10.5+1, 1)
bins_y = np.arange(-0.5, 10.5+1, 1)
plt.hist2d(data = df, x = 'height', y = 'weight', bins = [bins_x, bins_y])
plt.color();
```

```python
stats_df = df.drop(['height', 'weight', 'generation_id'], axis = 1)

#calculate correlations
corr = stats_df.corr()

#Heatmap
sb.heatmap(corr)
```

```python
sb.violinplot(data = df, X = 'generation_id', y = 'attack')
```
